```python
#IMPORTING THE REQUIRED LIBRARIES
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#IMPORTING THE DATASET
df = pd.read_excel("/content/globalterror.xlsx")
print("Dataset Imported successfully!")
```

```python
#DISPLAYING THE CONTENTS OF THE FIRST 8 ROWS OF THE DATA PROVIDED
df.head(8)
```

```python
#PLOTTING ATTACKING METHODS BY TERRORISTS
plt.figure(figsize=(10,5))
sns.countplot(df['attacktype1_txt'])
plt.xticks(rotation=90)
plt.title('ATTACKING METHODS BY TERRORISTS')
plt.xlabel('ATTACK TYPES')
plt.ylabel('ATTACKS')
plt.show()
```

```python
#PLOTTING FAVOURITE TARGETS
plt.figure(figsize=(20,10))
sns.countplot(df['targtype1_txt'])
plt.xticks(rotation=90)
plt.title('FAVOURITE TARGETS')
plt.xlabel('TARGET TYPE')
plt.ylabel('TARGET')
plt.show()
```

```python
#PLOTTING COUNTRIES ATTACKED
plt.figure(figsize=(20,8))
sns.barplot(x=df['country_txt'].value_counts()[:20].index,y=df['country_txt'].value_counts()[:20].values)
plt.title('COUNTRIES ATTACKED')
plt.xticks(rotation=40)
plt.xlabel('COUNTRY')
plt.ylabel('ATTACKS')
plt.show()
```

```python
#PLOTTING NO.OF ATTACKS PER YEAR
plt.figure(figsize=(12,8))
sns.countplot(df['iyear'])
plt.title('NUMBER OF ATTACKS PER YEAR')
plt.xticks(rotation=60)
plt.xlabel('YEAR')
plt.ylabel('NO.OF ATTACKS')
plt.show()
```

```python
#PLOTTING NUMBER OF ATTACKS IN EACH CITY
plt.figure(figsize=(20,10))
sns.barplot(x=df['city'].value_counts()[:20].index,y=df['city'].value_counts()[:20].values)
plt.title('NUMBER OF ATTACKS IN EACH CITY')
plt.xticks(rotation=40)
plt.xlabel('CITIES')
plt.ylabel('NO.OF ATTACKS')
plt.show()
```

```python
#PLOTTING TERRORIST ACTIVITIES IN EACH REGION
plt.figure(figsize=(15,5))
sns.countplot(df['region_txt'])
plt.title('TERRORIST ACTIVITIES IN EACH REGION')
plt.xticks(rotation=75)
plt.xlabel('REGIONS')
plt.ylabel('NO.OF ATTACKS')
plt.show()
```

```python
#PLOTTING FREQUENCY OF WEAPONS USED
x=df['weaptype1_txt'].value_counts()[:10].index
y=df['weaptype1_txt'].value_counts()[:10].values
sns.barplot(x,y)
plt.title('FREQUENCY OF WEAPONS USED')
plt.xlabel('WEAPONS USED')
plt.ylabel('FREQUENCY')
plt.xticks(rotation=90)
plt.show()
```

```python
#NUMBER OF FATALITIES PER YEAR
x=df['iyear']
y=df['nkill']
plt.scatter(x,y)
plt.title('NUMBER OF FATALITIES PER YEAR')
plt.xlabel('YEAR')
plt.ylabel('NUMBER OF FATALITIES')
plt.show()
```
